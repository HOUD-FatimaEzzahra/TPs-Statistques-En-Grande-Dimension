```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_table('wine.data', header=None, delimiter='\s+')
```

```python
data
```

```python
wines_cols=['Class_label','Alcohol','Malic acid','ash','Alcalinity of ash','Magnesium','Total phenols','Flavonoids','NonFlavonoid','Proanthocynins','Color intensity','Hue','00280/0D315 of diluter wines','Proline']
df=pd.read_csv('wine.data', header=None, names=wines_cols)
df
```

```python
df.info()
```

```python
df.drop(columns=['Class_label'],inplace=True)
```

```python
df
```

```python
from sklearn.pipeline import Pipeline
df_array=df.values
pipe=Pipeline([('scaler',StandardScaler()),
              ('acp',PCA(n_components=0.0))])
```

```python
pipe = Pipeline([('scaler', StandardScaler()), ('acp', PCA(n_components=2))])
```

```python
df_pca = pipe.fit_transform(df_array)
```

```python
df_pca_data = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
```
